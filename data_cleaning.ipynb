```python
import pandas as pd

# Caricamento del dataset
df = pd.read_csv('Gvibe_-_AI_reordered (1).csv')

# Creazione del dizionario per la conversione in Sistema Camelot (Chiave Armonica)
# La chiave è una tupla (Key, Mode) dove Mode: 0 = Minore (A), 1 = Maggiore (B)
camelot_mapping = {
    (0, 0): '5A',  (0, 1): '8B',
    (1, 0): '12A', (1, 1): '3B',
    (2, 0): '7A',  (2, 1): '10B',
    (3, 0): '2A',  (3, 1): '5B',
    (4, 0): '9A',  (4, 1): '12B',
    (5, 0): '4A',  (5, 1): '7B',
    (6, 0): '11A', (6, 1): '2B',
    (7, 0): '6A',  (7, 1): '9B',
    (8, 0): '1A',  (8, 1): '4B',
    (9, 0): '8A',  (9, 1): '11B',
    (10, 0): '3A', (10, 1): '6B',
    (11, 0): '10A',(11, 1): '1B'
}

# Applicazione della mappatura per convertire la colonna Key nella Chiave Armonica
df['Key'] = df.apply(lambda row: camelot_mapping.get((row['Key'], row['Mode']), str(row['Key'])), axis=1)

# Arrotondamento della colonna Tempo (bpm)
df['Tempo'] = df['Tempo'].round().astype(int)

# Selezione delle colonne richieste
colonne_da_tenere = [
    'Track Name', 
    'Artist Name(s)', 
    'Genres', 
    'Tempo', 
    'Key', 
    'Energy', 
    'Danceability', 
    'Popularity'
]
df_cleaned = df[colonne_da_tenere]

# Salvataggio del nuovo dataset modificato
df_cleaned.to_csv('track_library.csv', index=False)

# Visualizzazione delle prime righe
df_cleaned.head()
```
